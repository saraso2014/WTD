```python
import pandas as pd
import numpy as np
import random
import csv
from csv import writer
```

```python
List = ["Sara Was Here","Apple Pie"]
with open('data.csv', 'a') as f_object:  
    writer_object = writer(f_object)
  
    # Pass the list as an argument into
    # the writerow()
    writer_object.writerow(List)
  
    #Close the file object
    f_object.close()
```
